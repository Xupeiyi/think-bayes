```python
import numpy as np
from empiricaldist import Pmf
import matplotlib.pyplot as plt
from scipy.stats import binom
```

## 5.8 Exercises
### 5.8.1 Exercise 1

```python
hypos = np.arange(3, 2000)
prior = Pmf(1/len(hypos), hypos)

likelihood1 = binom.pmf(2, hypos, 1/365)
likelihood2 = binom.pmf(1, hypos-2, 1/364)
likelihood3 = binom.pmf(0, hypos-3, 1/363)

posterior = prior.copy()
posterior *= (likelihood1 * likelihood2 * likelihood3)
posterior.normalize()
posterior.plot()

print(posterior.mean())
print(posterior.max_prob())
print(posterior.prob_gt(1200))
```

### 5.8.2 Exercise 2

```python
n = np.arange(4, 11)
prior = Pmf(1 / len(n), n)
likelihood = (n - 1) / n**2

posterior = prior * likelihood
posterior.normalize()
prior.plot(label='prior')
posterior.plot(label='posterior')
```

### 5.8.3 Exercise 3

```python
hypos = np.arange(1, 4)
prior = Pmf(1, hypos)
likelihood = hypos
posterior = prior * likelihood
posterior.normalize()
display(posterior)

print(posterior.mean() / 2)
```
