```python
import pandas as pd
from fractions import Fraction
```

```python
table = pd.DataFrame(index=['Bowl 1', 'Bowl 2'])
table['prior'] = 1/2, 1/2
table['likelihood'] = 0.75, 0.5
table['unnormalized_posterior'] = table['prior'] * table['likelihood']
table['posterior'] = table['unnormalized_posterior'] / table['unnormalized_posterior'].sum()
print(table)
```

```python
table2 = pd.DataFrame(index=[6, 8, 12])
table2['prior'] = Fraction(1, 3)
table2['likelihood'] = Fraction(1, 6), Fraction(1, 8), Fraction(1, 12)

def update(table):
    """Compute the posterior probabilities."""
    table = table.copy()
    table['unnormalized_posterior'] = table['prior'] * table['likelihood']
    prob_data = table['unnormalized_posterior'].sum()
    table['posterior'] = table['unnormalized_posterior'] / prob_data
    return table

update(table2)
```

```python
table3 = pd.DataFrame(index=['Door 1', 'Door 2', 'Door 3'])
table3['prior'] = Fraction(1, 3)
table3['likelihood'] = Fraction(1, 2), 1, 0
update(table3)
```
